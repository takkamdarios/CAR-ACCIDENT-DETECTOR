```python
import os
import matplotlib.pyplot as plt
import cv2
import numpy as np

# Define the dataset directory
dataset_directory = "D:/ECOLEIT/4IA/Car Accident Detection Systeme/dataset/images"
```

```python
# Cell 2: Function to Load and Display a Batch of Images
def load_images_from_folder(folder, count=10):
    images = []
    for idx, filename in enumerate(os.listdir(folder)):
        if idx == count:
            break
        img = cv2.imread(os.path.join(folder, filename))
        if img is not None:
            images.append(img)
    return images

def display_images(images, titles=None, cmap=None):
    plt.figure(figsize=(20,10))
    for i, image in enumerate(images):
        plt.subplot(1, len(images), i+1)
        plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
        plt.axis('off')
        if titles is not None:
            plt.title(titles[i])
    plt.show()
```

```python
# Cell 3: Display a Batch of Images from the Dataset
sample_images = load_images_from_folder(dataset_directory)
display_images(sample_images)
```

```python
# Cell 4: Image Sizes Distribution
image_sizes = {}
for filename in os.listdir(dataset_directory):
    img = cv2.imread(os.path.join(dataset_directory, filename))
    if img is not None:
        size = img.shape[:2]
        if size not in image_sizes:
            image_sizes[size] = 0
        image_sizes[size] += 1

# Plotting the distribution of image sizes
sizes, counts = zip(*image_sizes.items())
plt.figure(figsize=(10, 5))
plt.bar(range(len(sizes)), counts, tick_label=[str(size) for size in sizes])
plt.xticks(rotation=90)
plt.xlabel('Image Sizes')
plt.ylabel('Frequency')
plt.title('Distribution of Image Sizes')
plt.show()
```

```python
# Cell 5: Check for Class Balance (If labels are available)
# Assuming a labels.csv file exists mapping filenames to labels
import pandas as pd

labels_csv_path = "D:/ECOLEIT/4IA/Car Accident Detection Systeme/dataset/labels/labels.csv"

labels_df = pd.read_csv(labels_csv_path)

# Visualizing class distribution
class_counts = labels_df['label'].value_counts()
plt.figure(figsize=(8, 4))
class_counts.plot(kind='bar')
plt.xlabel('Class')
plt.ylabel('Frequency')
plt.title('Class Distribution')
plt.show()
```
